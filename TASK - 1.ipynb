# Exploratory data analysis
### Exploratory Data Analysis or EDA is used to take deep understanding from the data. We try to find different patterns, relations, and anomalies in the data using some statistical graphs and other visualization techniques. 

## To make EDA work:

##### Get maximum insights from a data set
##### Uncover underlying structure
##### Extract important variables from the dataset
##### Detect outliers and anomalies(if any)
##### Test underlying assumptions
##### Determine the optimal factor settings

###### The main purpose of EDA is to detect any errors, outliers as well as to understand different patterns in the data.

##### Now let us Perform EDA in the given dataset prime.csv
